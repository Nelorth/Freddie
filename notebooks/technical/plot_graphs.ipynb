```python
plt.figure(figsize=(6, 6))
plt.subplot(2, 2, 1)
harry_plotter.plot_function(
    lambda x: np.where(np.abs(x) < 1e-10, x * np.nan, (x > 0) * 1  + (x < 0) * 0.005),
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{Heaviside}$",
    zorder=9999
)
harry_plotter.plot_function(
    lambda x: 0 * x,
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{Heaviside}'$",
    title="(a)"
)
plt.legend(loc="center left")
plt.subplot(2, 2, 2)
harry_plotter.plot_function(
    lambda x: 1 / (1 + np.exp(-x)),
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{expit}$",
    zorder=9999,
    color="tab:green"
)
harry_plotter.plot_function(
    lambda x: (1 / (1 + np.exp(-x))) * (1 - 1 / (1 + np.exp(-x))),
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{expit}'$",
    title="(b)",
    color="tab:red"
)
plt.legend()
plt.subplot(2, 2, 3)
harry_plotter.plot_function(
    lambda x: (x > 0) * x  + (x < 0) * 0.05,
    xlim=(-5, 5),
    ylim=(-5.5, 5.5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{ReLU}$",
    zorder=9999,
    color="tab:purple"
)
harry_plotter.plot_function(
    lambda x: np.where(np.abs(x) < 1e-10, x * np.nan, (x > 0) * 1),
    xlim=(-5, 5),
    ylim=(-5.5, 5.5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\mathrm{ReLU}'$",
    title="(d)",
    color="tab:pink"
)
plt.legend(loc="lower right")
plt.subplot(2, 2, 4)
harry_plotter.plot_function(
    lambda x: np.tanh(x),
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\tanh$",
    zorder=9999,
    color="tab:brown"
)
harry_plotter.plot_function(
    lambda x: 4 / (np.exp(x) + 1 / np.exp(x)) ** 2,
    xlim=(-5, 5),
    xlabel="$x$",
    ylabel="$y$",
    label="$\\tanh'$",
    title="(c)",
    color="tab:gray"
)
plt.legend()
plt.tight_layout()
plt.savefig(io.resolve_path(c.TEMP_DIR) / "func.pdf", bbox_inches="tight")
```
