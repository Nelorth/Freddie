# Event Boundary Visualization

```python
import pandas as pd
import harry_plotter
import utils

from constants import *
```

Retrieve labels

```python
df_labels = pd.read_csv(utils.resolve_path(LABEL_DIR, LABEL_FILE),
                        index_col=ORBIT_COL,
                        parse_dates=EVENT_COLS)
```

Load orbit $n$

```python
n = 1906
```

```python
df_orbit = pd.read_csv(utils.resolve_path(ORBIT_DIR, ORBIT_FILE(n)),
                       index_col=DATE_COL,
                       parse_dates=True)
```

Show event boundary timestamps

```python
df_labels.loc[n]
```

Plot magnetic flux density with event boundaries

```python
harry_plotter.plot_components_and_norm(
    df_orbit.index,
    (df_orbit["BX_MSO"],
    df_orbit["BY_MSO"],
    df_orbit["BZ_MSO"]),
    symbol="B",
    vlines=df_labels.loc[n],
    title=f"Orbit ${n}$",
    ylabel="magnetic flux density [\si{nT}]",
    outfile=utils.resolve_path(FIGURE_DIR, MESSENGER(n) + "_events.pdf")
)
```
