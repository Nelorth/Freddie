# Event Boundary Visualization

```python
import pandas as pd
import harry_plotter

from constants import *
from os.path import join
```

Retrieve labels

```python
df_labels = pd.read_csv(join(LABEL_DIR, LABEL_FILE),
                        index_col=ORBIT_COL,
                        parse_dates=EVENT_COLS)
```

Load orbit $n$

```python
n = 42
```

```python
df_orbit = pd.read_csv(join(ORBIT_DIR, ORBIT_FILE(n)),
                       index_col=DATE_COL,
                       parse_dates=True)
```

Plot magnetic flux density with event boundaries

```python
harry_plotter.plot_components_and_norm(
    df_orbit.index,
    (df_orbit["BX_MSO"],
    df_orbit["BY_MSO"],
    df_orbit["BZ_MSO"]),
    symbol="B",
    vlines=df_labels.loc[n]
)
```
